```python
import matplotlib.pyplot as plt
import numpy as np
```

# Estimating pi using LCG pPNG

```python
def LCG_seq(seed, n, a=1103515245, c=12345, m=32768):    
    x = seed
    list = []
    for i in range(n):
        x = (a*x + c)%m
        list.append(x/m)
    return list  
```

```python
def calculate_pi(seed1, seed2, N):
    inside_x, inisde_y, out_x, out_y= [], [], [], []
    x = LCG_seq(seed1, N)
    y = LCG_seq(seed2, N)
    for i, j in zip(x, y):
        mod = i**2 + j**2
        if mod <= 1:
            inside_x.append(i)
            inisde_y.append(j)
        elif mod >1:
            out_x.append(i)
            out_y.append(j)
    theta = np.linspace(0, np.pi/2, 1000)
    pi_val = 4*len(inside_x)/N
    plt.figure(figsize=(5, 5))
    plt.title(f'pi = {pi_val} for N = {N} ')        
    plt.plot(np.cos(theta), np.sin(theta), linestyle='-', color='black') 
    plt.scatter(inside_x, inisde_y, color = 'r', s = 1)       
    plt.scatter(out_x, out_y, color = 'b', s =1)       
    plt.show()
```

```python
calculate_pi(10, 11, 1000000)
```

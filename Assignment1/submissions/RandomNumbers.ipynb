# Question 1

```python
# function for generating sequence using logistic map

def iter_seq(c, seed,n = 10000):
    x = seed
    seq = [seed]
    for i in range(n-1):
        x = c*x*(1-x)
        seq.append(x)
    return seq
```

```python
N = 200
x0 = 0.1
c = 3.2
```

```python
#plotting x_i vs i

x = [i for i in range(N)]
y = iter_seq(c, x0, 200)
plt.plot(x, y)
plt.show()
```

 Correlation plots

```python
# function to return correlation map

def k_correlation(c, k = 5, fun = iter_seq, N = 10000):
    x, y = [], []
    list = fun(c, x0, N)
    for n in range(N-k-1):
        x.append(list[n])
        y.append(list[n + k])        
    # plt.scatter(x, y, s = 5)
    return x, y    
```

```python
fig, axs = plt.subplots(1, 3, figsize = (14, 5))

c1, c2, c3 = 3.9, 3.7, 3.8
k = 5
x1, y1 = k_correlation(c1, k)
x2, y2 = k_correlation(c2, k)
x3, y3 = k_correlation(c3, k)

axs[0].scatter(x1, y1, s = 4)
axs[0].set_title(f'c = {c1},  k = {k}')
axs[1].scatter(x2, y2, s = 4)
axs[1].set_title(f'c = {c2},  k = {k}')
axs[2].scatter(x3, y3, s = 4)
axs[2].set_title(f'c = {c3},  k = {k}')

fig.suptitle('Correlated plots')

plt.show()
```

```python
a = 1103515245
c = 12345
m = 32768
```

```python
def LCG_seq(seed, n, a = a, c = c, m = m):
    x = seed
    list = [x]
    for i in range(n):
        x = (a*x + c)%m
        list.append(x)
    return list    
```

```python
def lcg_k_correlation(k, N, seed, func = LCG_seq):
    x, y = [], []
    list = func(seed, N)
    for i in range(N - k -1):
        x.append(list[i])
        y.append(list[i+k])
    return x, y
```

```python
x, y = lcg_k_correlation(5, 1000, 10)

plt.scatter(x, y, s = 4)
plt.title('LCG correlation')
plt.show()
```

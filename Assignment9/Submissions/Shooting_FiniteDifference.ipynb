# Question1

```python
def x_dot(x, y):
    return y

def y_dot(x, y):
    return -0.01*(20-x)

shooting = Integrate.shooting

shooting(x_dot, y_dot, 12.87878665437314, [40, 200], [0, 10])   
```

```python
BVP = Integrate.BVP

sol = BVP(x_dot, y_dot, [10, 15], [40, 200], [0, 10])

x, y, t = sol['underestimate']
plt.plot(t, x)
x, y, t = sol['overestimate']
plt.plot(t, x)
x, y, t = sol['solution']
plt.plot(t, x)
plt.axhline(100, ls = '--')
plt.axvline(4.37, ls = '--', label = 'sol = 4.37 m')
plt.legend()
plt.show()
```

# Question1

```python

def y_dot(y, x):
    return (x+y)**2

def f(x):
    return np.tan(x+np.pi/4)-x

Runge_kutta = Integrate.Runge_kutta

x1, y1 = Runge_kutta(y_dot, [0, np.pi/5], 1, h = 0.1)
x2, y2 = Runge_kutta(y_dot, [0, np.pi/5], 1, h = 0.25)
x3, y3 = Runge_kutta(y_dot, [0, np.pi/5], 1, h = 0.45)

plt.plot(x1, y1, label = 'h = 0.1')   
plt.plot(x2, y2, label = 'h = 0.25')   
plt.plot(x3, y3, label = 'h = 0.45')   
y_true = [f(x) for x in x1]
plt.plot(x1, y_true, label = 'True solution')
plt.ylim(0, 20)
plt.legend()
plt.show()   
```

# Question2

```python
def x_dot(x, y):
    return y
def y_dot(x, y):
    return -x - 0.15*y

Runge_kutta = Integrate.Runge_kutta_2d

x_vals, y_vals, t_vals = Runge_kutta(x_dot, y_dot, [0, 40], [1, 0])
```

```python
plt.plot(t_vals, x_vals, label = 'x(t)')
plt.plot(t_vals, y_vals, label = 'y(t)')

E = [(x**2 + y**2)/2 for x,y in zip(x_vals, y_vals)]
plt.plot(t_vals, E, label = 'E(t)')
plt.grid(True)
plt.plot()
plt.title('Time Series')
plt.legend()
plt.show()
```

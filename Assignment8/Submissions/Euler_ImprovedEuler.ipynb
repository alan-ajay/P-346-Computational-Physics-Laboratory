# Question 1

```python
def y_dot(y, x):
    return y - x**2

def f(x):
    return x**2+2*x+2-2*np.exp(x)
```

```python
Euler_Forward = Integrate.Euler_Forward

x_vals, y_euler = Euler_Forward(y_dot, [0, 2], 0)
print(y_euler)
plt.plot(x_vals, y_euler, label = 'Approx solution')   
y_true = [f(x) for x in x_vals]
plt.plot(x_vals, y_true, label = 'True solution')
plt.legend()
plt.show()   

```

```python
Predictor_corrector = Integrate.Predictor_corrector

x_vals, y_pred = Predictor_corrector(y_dot, [0, 2], 0)
print(y_pred)
plt.plot(x_vals, y_pred, label = 'Approx solution')   
y_true = [f(x) for x in x_vals]
plt.plot(x_vals, y_true, label = 'True solution')
plt.legend()
plt.show()   

```

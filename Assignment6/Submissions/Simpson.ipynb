```python
# finding N for 10^-7 tolerance

mx1 = 2.29903 # max of fourth derivative of f1
mx2 = 4.20257 # max of fourth derivative of f2

error = 1e-7

nm1 = int(np.ceil(np.sqrt((2/(24*error)))))
nm2 = int(np.ceil(np.sqrt((((np.pi/2)**3)*mx1)/(24*error))))

ns1 = int(np.ceil((24/(180*error))**(1/4)))
ns2 = int(np.ceil((((np.pi/2)**5)*mx2/(180*error))**(1/4)))+1

print(nm1, nm2, ns1, ns2)
```

```python
#importing functions from Integrate class
midpoint = Integrate.Midpoint
simpson = Integrate.Simpson

def f1(x):
    return 1/x

def f2(x):
    return x* np.cos(x)



m1 = midpoint(f1, 1, 2, nm1)
m2 = midpoint(f2, 0, np.pi/2, nm2)
s1 = simpson(f1, 1, 2, ns1)
s2 = simpson(f2, 0, np.pi/2, ns2)

print('f1 = 1/x')
print(f'ANALYTICAL RESULT:  {0.69314718}\nmidpoint for f1:    {m1}\nsimpson for f1:     {s1}')
print('--------------')
print('f2 = xcosx')
print(f'ANALYTICAL RESULT:  {(np.pi/2)-1}\nmidpoint for f2:    {m2}\nsimpson for f2:     {s2}')
```

# Question2

```python
montecarlo = Integrate.Montecarlo

def f(x):
    return np.sin(x)**2


true_val = 0.545351

N = 100000
x_vals = [i for i in range(1, N, 1000)]
y_vals = [montecarlo(f, -1, 1, n)[0] for n in x_vals]

plt.axhline(true_val, label=f'Analytical Value   = {true_val}')
plt.plot(x_vals, y_vals, color = 'r', label=f'Montecarlo result = {round(y_vals[-1],6)}')
plt.xlabel('N')
plt.ylabel(f'Integral value')
plt.title(f'Integral value vs N for Montecarlo integration(N = {N})')
plt.legend()
plt.grid('True')
# plt.ylim(0.53, 0.55)
plt.show()
```

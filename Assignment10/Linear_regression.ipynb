```python
least_square_fit = LinReg.least_square_fit
ln_tuple = LinReg.ln_tuple

data = [(2.5,13.0),(3.5,11.0),(5.0,8.5),(6.0,8.2),(7.5,7.0),(10.0,6.2),(12.5,5.2),(15.0,4.8),(17.5,4.6),(20.5,4.3)]
data1 = ln_tuple(data, 1) #took log for both
data2 = ln_tuple(data, 2) # took log for y
s = [1 for _ in range(len(data))] # set of sigmai's

sol1, sol2 = least_square_fit(data1, s), least_square_fit(data2, s)


print('(a1, a2, r^2) = \n', sol1, '\n', sol2)

print(f'\n power law: r**2 = {sol1[2]} \n exp: r**2 = {sol2[2]}')
```

```python
def f1(x):#power law
    a1, a2 =  np.exp(sol1[0]), sol1[1]
    return a1*x**a2

def f2(x):#exp
    a1, a2 =  np.exp(sol1[0]), -sol1[1]
    return a1*np.exp(-a2*x)

n = len(data)
x = np.array([data[i][0] for i in range(n)])
y = [data[i][1] for i in range(n)]

plt.plot(x, y, label= 'data')
plt.plot(x, f1(x), label = 'power law')
plt.plot(x, f2(x), label = 'exp')
plt.ylim(0, 20)
plt.legend()
plt.show()
```
